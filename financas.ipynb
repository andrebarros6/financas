```python
cont_selected = cont[["Data de Emissão", "Nome do Adquirente", "Total do Documento (em euros)"]]
cont_selected.head()
```

```python
cont_selected["Mês"] = pd.to_datetime(cont_selected["Data de Emissão"], dayfirst=True).dt.month
```

```python
cont_selected.head(10)
```

```python
cont_selected.info()
```

```python
cont_selected["Total do Documento (em euros)"] = cont_selected["Total do Documento (em euros)"].str.replace(".", "")
cont_selected["Total do Documento (em euros)"] = cont_selected["Total do Documento (em euros)"].str.replace(",", ".").astype(float)
```

```python
total_2025 = cont_selected["Total do Documento (em euros)"].sum()
total_2025
```

```python
total_mes = cont_selected.groupby("Mês")["Total do Documento (em euros)"].sum()
#total_mes["month"] = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
# total_mes = pd.DataFrame(total_mes).set_index("month")
```

```python
plt.bar(total_mes.index, total_mes.values)
plt.show()
```

```python
total_nome = cont_selected.groupby("Nome do Adquirente")["Total do Documento (em euros)"].sum()
# total_mes["month"] = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
# total_mes = pd.DataFrame(total_mes).set_index("month")
```

```python
plt.bar(total_nome.index, total_nome.values)
plt.xticks(rotation=90)
plt.show()
```

```python
nome_por_mes = cont_selected.groupby(["Nome do Adquirente","Mês"])["Total do Documento (em euros)"].sum()
mes_por_nome = cont_selected.groupby(["Mês","Nome do Adquirente"])["Total do Documento (em euros)"].sum()
```

```python
mes_por_nome_df = pd.DataFrame(mes_por_nome)
mes_por_nome_df.reset_index()
```

```python
mes_por_nome_df.groupby(['Mês', 'Nome do Adquirente']).size().unstack().plot(kind='bar', stacked=True)
# isto dá o count de recibos passados - como já está somado, cada clínica só aparece uma vez
# pivot it
```

```python
df_pivot = mes_por_nome_df.pivot_table(
    values = "Total do Documento (em euros)",
    index = "Mês",
    columns = "Nome do Adquirente",
    aggfunc = 'sum'
)
```

```python
df_pivot
```

```python
df_pivot.plot(kind='bar', stacked=True)
```
